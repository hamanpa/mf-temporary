```python
# Excitatory cells (RS) data

rs_data = np.load("../projects/01_fitting_tf/zerlaut2018_data/RS-cell_CONFIG1.npy", allow_pickle=True)

nu_out_mean = rs_data[0]  # (10, 10) array, Mean output firing rate (Hz)
nu_out_std = rs_data[1]  # (10, 10) array, STD firing rate (Hz)
nu_exc = rs_data[2]  # (10, 10) array, Excitatory input (Hz)
nu_inh = np.stack([rs_data[3]]*10, axis=1)  # (10, 10) array, Inhibitory input (Hz)
params = rs_data[4]  # dict with params (SI units)

# (row, col) == (inh_idx, exc_idx)
for inh_idx, inh_input in enumerate(nu_inh[:,0]):
    plt.errorbar(nu_exc[inh_idx], nu_out_mean[inh_idx], yerr=nu_out_std[inh_idx], fmt="o--", label=f"nu_inh={inh_input:.1f} Hz")
plt.legend()
plt.show()

plt.contourf(nu_exc, nu_inh, nu_out_mean)
plt.colorbar()
plt.xlabel(r"$\nu_{exc}\,[Hz]$")
plt.ylabel(r"$\nu_{inh}\,[Hz]$")
plt.show()
```

```python
# Excitatory cells (RS) data I generated using the code from repo (had to correct some bugs in the code, see other notebook)

rs_data_generated = np.load("../projects/01_fitting_tf/zerlaut2018_data_generated/RS-cell_CONFIG1.npy", allow_pickle=True)

nu_out_mean_generated = rs_data_generated[0]  # (10, 10) array, Mean output firing rate (Hz)
nu_out_std_generated = rs_data_generated[1]  # (10, 10) array, STD firing rate (Hz)
nu_exc_generated = rs_data_generated[2]  # (10, 10) array, Excitatory input (Hz)
nu_inh_generated = np.stack([rs_data_generated[3]]*10, axis=1)  # (10, 10) array, Inhibitory input (Hz)
params_generated = rs_data_generated[4]  # dict with params (SI units)

# (row, col) == (inh_idx, exc_idx)
for inh_idx, inh_input in enumerate(nu_inh_generated[:,0]):
    plt.errorbar(nu_exc_generated[inh_idx], nu_out_mean_generated[inh_idx], yerr=nu_out_std_generated[inh_idx], fmt="o--", label=f"nu_inh={inh_input:.1f} Hz")
plt.legend()
plt.show()

plt.contourf(nu_exc_generated, nu_inh_generated, nu_out_mean_generated)
plt.colorbar()
plt.xlabel(r"$\nu_{exc}\,[Hz]$")
plt.ylabel(r"$\nu_{inh}\,[Hz]$")
plt.show()
```

# Comparison of neuron data with NEST generated ones

```python
# Zerlaut neural data (excitatory neuron)

nu_out_mean = rs_data[0]  # (10, 10) array, Mean output firing rate (Hz)
nu_out_std = rs_data[1]  # (10, 10) array, STD firing rate (Hz)
nu_exc = rs_data[2]  # (10, 10) array, Excitatory input (Hz)
nu_inh = np.stack([rs_data[3]]*10, axis=1)  # (10, 10) array, Inhibitory input (Hz)
params = rs_data[4]  # dict with params (SI units)

for inh_idx, inh_input in enumerate(nu_inh[:,0]):
    plt.errorbar(nu_exc[inh_idx], nu_out_mean[inh_idx], yerr=nu_out_std[inh_idx], fmt="o--", label=f"nu_inh={inh_input:.1f} Hz")

plt.legend()
```

```python
# MFT generated neural data (excitatory neuron)

with open("./data/divolo-static_exc_neuron_16x16_grid-irregular.pickle", "rb") as my_file:
    my_data = pickle.load(my_file)

for inh_idx, inh_input in enumerate(my_data.nu_i[0]):
    plt.plot(my_data.nu_e[:,inh_idx], my_data.out_rate_mean[:,inh_idx], "o--", label=f"nu_inh={inh_input:.1f} Hz")
plt.legend()
```

```python
# Rough comparison of the two datasets (Zerlaut and MFT generated) for excitatory neuron


step = 2
for inh_idx, inh_input in enumerate(nu_inh[:,0]):
    if inh_idx % step != 0:
        continue
    plt.errorbar(nu_exc[inh_idx], nu_out_mean[inh_idx], yerr=nu_out_std[inh_idx], color='black', fmt="x--", label=f"Zerlaut nu_inh={inh_input:.1f} Hz")


with open("./data/divolo-static_exc_neuron_16x16_grid-irregular.pickle", "rb") as my_file:
    my_data = pickle.load(my_file)


for inh_idx, inh_input in enumerate(my_data.nu_i[0]):
    if inh_input > 21:
        continue
    plt.plot(my_data.nu_e[:,inh_idx], my_data.out_rate_mean[:,inh_idx], "o:", color='red',  label=f"MFT nu_inh={inh_input:.1f} Hz")
plt.legend()

plt.ylim(0, 25)
# plt.xlim(0, 25)
plt.xlim(0, 7)
plt.tick_params(axis='y', right=True, labelright=True) 
```
